# Subpart 5: Weekly changes in number of hotspots per state

```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

Reading the csv file and filtering the cases

```python
cols=['detected_district','status_change_date','current_status','detected_state']
ind_cases=pd.read_csv('IndividualDetails.csv', parse_dates =['status_change_date'], dayfirst=True,usecols=cols)
ind_cases =ind_cases.dropna()
ind_cases =ind_cases[ind_cases['current_status']=="Hospitalized"]
```

Determining the week 1 hotspots

```python
week1 = '2020-03-27'
df =ind_cases[ind_cases.status_change_date <=week1]
df['Active']=df.groupby('detected_district')['detected_district'].transform('count')
df=df[['detected_district','Active','detected_state']]
df=df.drop_duplicates()
df['Hotspots1']=df.groupby('detected_state')['detected_state'].transform('count')
df=df[['detected_state','Hotspots1']]
df=df.drop_duplicates()
df
```

Selecting and displaying the states with the maximum hotspots in the first week

```python
week1_hotspots=df.sort_values('Hotspots1', ascending=False).reset_index(drop=True)
week1_hotspots.head()
```

Determining the week 2 hotspots

```python
week2='2020-04-03'
df =ind_cases[ind_cases.status_change_date <=week2]
df['Active']=df.groupby('detected_district')['detected_district'].transform('count')
df=df[['detected_district','Active','detected_state']]
df=df.drop_duplicates()
df['Hotspots2']=df.groupby('detected_state')['detected_state'].transform('count')
df=df[['detected_state','Hotspots2']]
df=df.drop_duplicates()
df
```

Selecting and displaying the states with the maximum hotspots in the second week

```python
week2_hotspots=df.sort_values('Hotspots2', ascending=False).reset_index(drop=True)
week2_hotspots.head()
```

Determining the week 3 hotspots

```python
week3='2020-04-17'
df =ind_cases[ind_cases.status_change_date <=week3]
df=df.sort_values(by='status_change_date',ascending=False)
df['Active']=df.groupby('detected_district')['detected_district'].transform('count')
df=df.drop_duplicates()
df['Hotspots3']=df.groupby('detected_state')['detected_state'].transform('count')
df=df[['detected_state','Hotspots3']]
df=df.drop_duplicates()
```

Selecting and displaying the states with the maximum hotspots in the third week

```python
week3_hotspots=df.sort_values('Hotspots3', ascending=False).reset_index(drop=True)
week3_hotspots.head()
```

Weekwise hotspots for all the states

```python
week_wise=week1_hotspots.merge(week2_hotspots)
week_wise=week_wise.merge(week3_hotspots)
week_wise
```

Calculating and displaying the states which have the maximum increase in hotspots between the first and second week

```python
#max increase in week 1-2
week_wise_dif1=week_wise[['detected_state']]
week_wise_dif1['Increase']=week_wise['Hotspots2']-week_wise['Hotspots1']
week_wise_dif1=week_wise_dif1.sort_values('Increase',ascending=False).reset_index(drop=True)
week_wise_dif1=week_wise_dif1.rename(columns={'detected_state': 'States with highest increase from W1 to W2'})
week_wise_dif1.head()
```

Calculating and displaying the states which have the maximum increase in hotspots between the second and third week

```python
week_wise_dif2=week_wise[['detected_state']]
week_wise_dif2['Increase']=week_wise['Hotspots3']-week_wise['Hotspots2']
week_wise_dif2=week_wise_dif2.sort_values('Increase',ascending=False).reset_index(drop=True)
week_wise_dif2=week_wise_dif2.rename(columns={'detected_state': 'States with highest increase from W2 to W3'})
week_wise_dif2.head()
```

Calculating and displaying the states which have the maximum decrease in hotspots between the first and second week

```python
week_wise_dec1=week_wise[['detected_state']]
week_wise_dec1['Decrease']=week_wise['Hotspots1']-week_wise['Hotspots2']
week_wise_dec1=week_wise_dec1.sort_values('Decrease',ascending=False).reset_index(drop=True)
week_wise_dec1=week_wise_dec1.rename(columns={'detected_state': 'States with highest decrease from W1 to W2'})
if week_wise_dec1['Decrease'][0] ==0:
    print("There is no state that shows a decrease of hotspots between Week 1 and Week 2")
else:
    week1_dec=week_wise_dec1.loc[week_wise_dec1['Decrease'] > 0]
    week1_dec.head()  
```

Calculating and displaying the states which have the maximum decrease in hotspots between the second and third week

```python
week_wise_dec2=week_wise[['detected_state']]
week_wise_dec2['Decrease']=week_wise['Hotspots1']-week_wise['Hotspots2']
week_wise_dec2=week_wise_dec2.sort_values('Decrease',ascending=False).reset_index(drop=True)
week_wise_dec2=week_wise_dec2.rename(columns={'detected_state': 'States with highest decrease from W1 to W2'})
if week_wise_dec2['Decrease'][0] ==0:
    print("There is no state that shows a decrease of hotspots between Week 2 and Week 3")
else:
    week2_dec=week_wise_dec2.loc[week_wise_dec2['Decrease'] > 0]
    week2_dec.head()  
```
